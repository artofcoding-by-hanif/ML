Mean Squared Error (MSE) or Cost Function
Gradient Decent -> Is an algorithm that finds best fit line for given traininf

y_predicted = mX + b
MSE = 1/n SUM (y_input - y_predicted)2 

Derivative
Partial Derivative - Gives the slope
Learning rate - constant

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_curr = b_curr = 0
    rate = 0.08
    n = len(x)
    plt.scatter(x,y)
    for i in range(1000):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        print (m_curr,b_curr, cost, i)
        plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted)) # M derivative
        yd = -(2/n)*sum(y-y_predicted) # Y derivative
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
        
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
